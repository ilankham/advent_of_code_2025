# Set puzzle parameters and create AoC Session

```python
# set puzzle parameters
PUZZLE_DAY = 1
PUZZLE_YEAR = 2025

# import from local packages
from aoc_solver import AoCSession, AoCSolver, AoCTester
AoC_SESSION = AoCSession.from_file()
```

# Import additional packages

```python
# import from standard library packages
from math import floor
from typing import Tuple

# import from third-party packages
from tqdm.notebook import tqdm
```

# Create solver class and instance

```python
class Solver(AoCSolver):

    def solve_part1(self, data: Tuple[str]) -> int:
        lock_position = 50
        number_of_instances_of_zero = 0
        for rotation in tqdm(data):
            direction = rotation[0]
            steps = int(rotation[1:])
            if direction == 'L':
                lock_position = (lock_position - steps) % 100
            elif direction == 'R':
                lock_position = (lock_position + steps) % 100
            else:
                assert 'Undefined direction!'
            if lock_position == 0:
                number_of_instances_of_zero += 1
        return number_of_instances_of_zero

    def solve_part2(self, data: Tuple[str]) -> int:
        lock_position = 50
        number_of_instances_of_zero = 0
        for rotation in tqdm(data):
            direction = rotation[0]
            steps = int(rotation[1:])

            # account for extra rotations when steps > 100
            number_of_instances_of_zero += floor(steps/100)

            if direction == 'L':
                # account for rotating past 0 when not starting at 0
                if lock_position != 0 and steps % 100 >= lock_position:
                    number_of_instances_of_zero += 1
                lock_position = (lock_position - steps) % 100
            elif direction == 'R':
                # account for rotating past 0 when not starting at 0
                if lock_position != 0 and steps % 100 >= (100 - lock_position):
                    number_of_instances_of_zero += 1
                lock_position = (lock_position + steps) % 100
            else:
                assert 'Undefined direction!'
        return number_of_instances_of_zero
```

```python
solver = Solver(PUZZLE_YEAR, PUZZLE_DAY, AoC_SESSION)
```

# Build part 1 test case(s)

```python
puzzle_instructions = solver.puzzle_instructions

part1_test_input = solver.get_value_after('the following rotations:').create_tuple(str,'\n')
print(f'{part1_test_input=}\n')

part1_test_output = solver.get_value_after('in this example is ').as_int
print(f'{part1_test_output=}\n')
```

```python
part_1_tester = AoCTester()
part_1_tester.add_test_case(part1_test_input, part1_test_output)
```

```python
%%time
part_1_tester.run_tests(solver.solve_part1)
```

# Determine part 1 solution

```python
%%time
puzzle_input = solver.puzzle_input.create_tuple(str,'\n')
part1_solution = solver.solve_part1(puzzle_input)
print(f'{part1_solution=}\n')
```
