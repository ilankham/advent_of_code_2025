# Set puzzle parameters and create AoC Session

```python
# set puzzle parameters
PUZZLE_DAY = 3
PUZZLE_YEAR = 2025

# import from local packages
from aoc_solver import AoCSession, AoCSolver, AoCTester
AoC_SESSION = AoCSession.from_file()
```

# Import additional packages

```python
# import from standard library packages
from typing import Tuple

# import from third-party packages
from tqdm.notebook import tqdm
```

# Create solver class and instance

```python
class Solver(AoCSolver):
    def solve_part1(self, data: Tuple[str]) -> int:
        total_output_joltage = 0
        for battery_bank in tqdm(data):
            max_prefix = max(
                int(i) for i in battery_bank[:-1]
            )
            max_prefix_index = battery_bank.index(str(max_prefix))
            max_suffix = max(
                int(i) for i in battery_bank[max_prefix_index+1:]
            )
            joltage = max_prefix * 10 + max_suffix
            total_output_joltage += joltage
        return total_output_joltage

    def solve_part2(self, data: Tuple[str]) -> int:
        total_output_joltage = 0
        for battery_bank in tqdm(data):
            max_prefix = max(
                int(i) for i in battery_bank[:-11]
            )
            current_index = battery_bank.index(str(max_prefix))
            joltage = max_prefix * 10**11
            for i in range(1,11):
                current_segment = battery_bank[current_index+1:-11+i]
                next_digit = max(
                    int(i) for i in current_segment
                )
                current_index = battery_bank.index(str(next_digit),current_index+1)
                joltage += next_digit * 10**(11-i)
            last_digit = max(
                int(i) for i in battery_bank[current_index+1:]
            )
            joltage += last_digit
            total_output_joltage += joltage
        return total_output_joltage
```

```python
solver = Solver(PUZZLE_YEAR, PUZZLE_DAY, AoC_SESSION)
```

# Build part 1 test case(s)
